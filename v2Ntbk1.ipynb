## Exploratory Data Analysis

In this version we have added a new feature [Kenya Indicator Foreign Exchange Reserves] which contains historical data from october 2016 
to october 2024.


##### Plot
- plot of Interest Rate vs Kenya Foreign Exchange Reserves:
  - There was a weak positive correlation.
- plot of Mean Value vs Kenya Foreign Exchange Reserves:
  - There was no notable correlation between the two.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

```

```python
data = pd.read_csv("data.csv", index_col='Date', parse_dates=True)
```

##### Check the correlation between three features; interest rate, mean value & kenya reserves

- 1.1 mean vaue vs kenya reserves
- 1.2 interest rate vs kenya reserves

```python
#Plotting
#plot 1

fig, ax1 = plt.subplots(figsize=(12, 5))

ax1.plot(data.index, data['Mean-value'], label='Mean-value', color='green')
ax1.set_xlabel('Date')
ax1.set_ylabel('Currency Price (KSH)', color='green')
ax1.tick_params(axis='y', labelcolor='green')

ax2 = ax1.twinx()  # Create a twin axis sharing the same x-axis
ax2.plot(data.index, data['kenya_reserves'], label='Kenya Reserves', color='red')
ax2.set_ylabel('Kenya Foreign Exchange Reserves', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Currency Price(ksh) and Kenya Foreign Exchange Reserves over time')
plt.grid()

ax1.legend(loc='upper left')
ax2.legend(loc='upper center')

plt.show()


#plot2
fig, ax1 = plt.subplots(figsize=(12, 5))

ax1.plot(data.index, data['Interest-rate'], label='Interest-rate', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Interest Rate', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()  # Create a twin axis sharing the same x-axis
ax2.plot(data.index, data['kenya_reserves'], label='Kenya Reserves', color='red')
ax2.set_ylabel('Kenya Foreign Exchange Reserves', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('US Interest Rates and Kenya Foreign Exchange Reserves over time')
plt.grid()

ax1.legend(loc='upper left')
ax2.legend(loc='upper center')

plt.show()
```

##### Observation

fig1 : no notable correlation.

fig2 : A weak positive correlation between the two features.
